# ROC - AUC #

ROC Curve : Sensitivity vs 1-Specificity





FPR -> False Positive Rate (1-Specificity)
TPR -> True Positive Rate  (Sensitivity)



Area under the Curve (AUC), Bigger the area under the curve better is the model

roc_curve : Calculates the valies required for drawing ROC curve

roc_auc_score : Calculates the area under the ROC curve

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
fpr, tpr, threshold = roc_curve(df['y_test'],df['yprob_1'])
```

```python
plt.plot(fpr,tpr)
plt.xlabel("1-Specificity")
plt.ylabel("Sensitivity")
plt.show()
print(roc_auc_score(df['y_test'],df['yprob_1']))
```

### 2ND MODEL ###

```python
fpr, tpr, threshold = roc_curve(df['y_test'],df['yprob_2'])
```

```python
plt.plot(fpr,tpr)
plt.xlabel("1-Specificity")
plt.ylabel("Sensitivity")
plt.show()
print(roc_auc_score(df['y_test'],df['yprob_2']))
```

From the above two charts it is clearly visible that the 2nd model is better since it has more area under curve **(0.7916666666666666 vs 0.6666666666666666)**
