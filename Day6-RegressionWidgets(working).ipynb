# CORRELATION #

- To find association between two variables

Two types of Correlation :

- **Linear** -> X and Y (Highest Degree is 1)
- **Non-Linear** -> X^2 and Y (Highest Degree is more than 1)
--------------------------------------------------------------------------------------------------------------------------------
- **Linear**: Corresponding to a unit change in one variable, there is a constant
    change in the other variable.
- **Non-Linear**: Corresponding to a unit change in one variable, the other variable doesn’t change at a constant rate but it    changes at a fluctuating rate.

**Scatter plots** and **correlation co-efficients** helps us visualise the relation between two variables



### Karl Pearson's Coefficient ###


Karl Pearson's coefficient of correlation is defined as a linear correlation coefficient that falls in the value range of -1 to +1.

Assumptions of Linear Regression :

- Normality : Errors are Normally Distributed with mean 0

- Independence : Errors are independent

- Linearity : Mean of Y is linearly related to X's

- Homoscedasticity : Errors have constant variance


This is very difficult to ensure in real life. So simple regression is generally not used.

## Regularized Regression ##

Regularized Regression Techniques :

- Ridge
- Lasso
- ElasticNet (Combination of previous two)

```python
from sklearn.linear_model import Ridge, Lasso , ElasticNet
boston = pd.read_csv("Boston.csv")
boston.head(10)
```

```python
X = boston.drop('medv',axis=1)
y = boston['medv']
```

### Ridge ###

```python
params={'alpha':[0.01,0.1,0.5,1,2,3,6,10]}
```

```python
ridge = Ridge()
kfold=KFold(n_splits=5, shuffle=True, random_state=2023)
gcv = GridSearchCV(ridge, param_grid=params, cv=kfold, scoring='r2') #r2 is used since its a regression problem
gcv.fit(X,y)

print(gcv.best_params_)
print(gcv.best_score_)
pd.DataFrame(gcv.cv_results_)
```

```python
best=gcv.best_estimator_
print(best.coef_)
```

```python
params={'RIDGE__alpha':[0.01,0.1,0.5,1,2,3,6,10]}
```

```python
#Ridge with scaling
ridge = Ridge()
kfold=KFold(n_splits=5, shuffle=True, random_state=2023)
scaler=StandardScaler()
pipe=Pipeline([('STD',scaler),('RIDGE',ridge)])
gcv = GridSearchCV(pipe, param_grid=params, cv=kfold, scoring='r2') #r2 is used since its a regression problem
gcv.fit(X,y)

print(gcv.best_params_)
print(gcv.best_score_)
pd.DataFrame(gcv.cv_results_)
```

```python
best=gcv.best_estimator_.named_steps['RIDGE']
print(best.coef_)
```

### ElasticNet ###

```python
params={'alpha':[0.01,0.1,0.5,1,2,3,6,10],'l1_ratio':[0,0.25,0.5,0.75,1]}
```

### alpha is equal to lamda that we used in the equation of ridge and lasso

#### What is L1 ratio in elastic net?
This is called the ElasticNet mixing parameter. Its range is 0 < = l1_ratio < = 1. If l1_ratio = 1, the penalty would be L1 penalty. If the value of l1 ratio is between 0 and 1, the penalty would be the combination of L1 and L2.

```python
elastic = ElasticNet()
kfold=KFold(n_splits=5, shuffle=True, random_state=2023)
gcv = GridSearchCV(elastic, param_grid=params, cv=kfold, scoring='r2') 
gcv.fit(X,y)

print(gcv.best_params_)
print(gcv.best_score_)
pd.DataFrame(gcv.cv_results_)
```

```python
poly = PolynomialFeatures(degree=2)
X_poly=poly.fit_transform(X)
kfold=KFold(n_splits=5, shuffle=True, random_state=2023)
res=cross_val_score(lr, X_poly, y, scoring = 'r2', cv=kfold)
print(poly.get_feature_names_out())
# get-feature_names_out : return the names of column
res.mean()
```

```python
for i in range(1,5):
    poly = PolynomialFeatures(degree=i)
    X_poly=poly.fit_transform(X)
    kfold=KFold(n_splits=5, shuffle=True, random_state=2023)
    res=cross_val_score(lr, X_poly, y, scoring = 'r2', cv=kfold)
    #print(poly.get_feature_names_out())
    print(res.mean())
```

```python
# without using gcv we passed degree value inside the class PloynomailFeature() 
# But with gcv we use params so passing degree to params and passing pipe which having both Poly + linear to the gcv

params={'POLY__degree':[1,2,3,4]}
poly = PolynomialFeatures()
lr=LinearRegression()
kfold=KFold(n_splits=5, shuffle=True, random_state=2023)
pipe=Pipeline([('POLY',poly),('LIN',lr)])
gcv = GridSearchCV(pipe, param_grid=params, cv=kfold, scoring='r2')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)

# here in gcv.fit we are using x instead of X_poly beac x_poly was used to tranform the degree externally ,
# now here  pipe automatically doing it
```

```python
ridge = Ridge()
kfold=KFold(n_splits=5, shuffle=True, random_state=2023)
scaler=StandardScaler()
pipe=Pipeline([('STD',scaler),('RIDGE',ridge)])
gcv = GridSearchCV(pipe, param_grid=params, cv=kfold, scoring='r2') #r2 is used since its a regression problem
gcv.fit(X,y)

print(gcv.best_params_)
print(gcv.best_score_)
pd.DataFrame(gcv.cv_results_)
```
